### Using `if/else` or `try-ecept` Best Practice Recommendation:

While both methods are valid and have their uses, the `try-except` approach is often more aligned with Pythonic principles of "it's easier to ask for forgiveness than permission" (EAFP). This principle suggests that you should assume the best and deal with problems if they actually occur.
In this case, the `try-except` version is preferable because it cleanly handles the types of errors that division might produce, such as `ZeroDivisionError` and `TypeError`, without needing multiple conditional checks beforehand. It's efficient, clean, and adheres to common Python practices.



### Using the`try-ecept` `else` Block:

*   **Clarity and Separation**: The `else` block in a `try-except` structure runs only if the code inside the `try` block did not raise an exception. This clearly separates the code that might cause exceptions from the code that should run only if no exceptions were raised. It's useful for keeping the error handling cleanly separated from normal execution.
*   **Safety**: It prevents accidentally catching exceptions that weren't meant to be caught by the `try` block, which might happen if you put too much code in the `try` block.

### Best Practice:

*   **Use `else` when**: You have additional code that should only execute if the `try` block did not raise an exception, and this code itself should not be covered by the preceding `except` clauses.
*   **Use `return` in `try` when**: The function needs to output a result directly from the attempted operation, and you want to handle the success case as soon as the operation completes successfully.

In practice, the choice between using an `else` block or a `return` statement inside the `try` block often comes down to the specific requirements of your application and which option makes the code more readable and maintainable.

Self exercice of playing with the class code so it can draw multiple circles instead of one of a time, giving the distance between each circle of its radius sum:

````python
# My Code
# Import the library
import importlib.util
if importlib.util.find_spec("matplotlib") is None:
    %pip install matplotlib
    
import matplotlib.pyplot as plt
%matplotlib inline

# Create a class Circle
class Circle(object):
    
    # Constructor
    def __init__(self, radius=3, color='blue', center=(0, 0)):
        self.radius = radius
        self.color = color
        self.center = center
    
    # Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return self.radius
    
    # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle(self.center, radius=self.radius, fc=self.color))
        plt.axis('scaled')


# Create a list of circles
circles = []

# Create N circles and add them to the list
N = 3  # Number of circles
spacing = 0  # Initial spacing
for i in range(N):
    radius = 1 + i  # Increase radius for each circle
    color = 'blue' if i % 2 == 0 else 'red'  # Alternate colors
    center = (radius + spacing, 0)  # Adjust x-coordinate for each circle
    circles.append(Circle(radius, color, center))
    spacing += 2 * radius  # Update spacing for next circle

# Draw circles side by side
for circle in circles:
    circle.drawCircle()

plt.show()
````
